```python
geometry = list(zip(df['pickup_longitude'], df['pickup_latitude']))
```

```python
geometry = [Point(x) for x in geometry]
```

```python
crs = {'init': 'epsg:4326'}
df = df.drop(['pickup_longitude','pickup_latitude'], axis=1)
taxi_gdf = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
```

```python
zones = zones.to_crs({'init': 'epsg:4326'})
```

```python
taxi_results = gpd.sjoin(taxi_gdf, zones, how='left', op='intersects')
```

```python
taxi_results.to_csv('taxi_output.csv', index=False)
```
